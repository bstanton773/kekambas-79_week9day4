### Data Visualization <br>
<p>Data visualization is one of the most powerful tools an analyst has for two main reasons. Firstly, it is unrivalled in it's ability to guide the analysts hand in determining 'what to look at next'. Often a visual is revealing of patterns in the data that are not easily discernable by just looking at dataframes. Secondly, they are an analysts greatest communication tool. Very often professional analysts need to present their results to groups of people responsible for taking action based on what the data says. Visuals can tell your story much better then raw numbers, so let's talk a little bit about how to make effective visuals that make a point.</p>

### Different Types of Charts <br>
<ul>
    <li><b>Line Chart:</b> Exploring data over time</li>
    <li><b>Bar Chart:</b> Comparing categories of data and tracks changes over time</li>
    <li><b>Pie Chart:</b> Explore parts of a whole, i.e. fractions</li>
    <li><b>Scatter Plot:</b> Similar to line charts, track correlations between two categories</li>
    <li><b>Histogram:</b> Unrelated from bar charts, show distribution of variables</li>
    <li><b>Candlestick Chart:</b> Used a lot in financial sector, i.e. can compare a stock over a period of time</li>
    <li><b>Box Chart:</b> Looks identical to candlestick charts, compares minimum, 1st, medium, 3rd quartiles, and max values</li>
</ul>

<p>Depending on what you need to accomplish in conceptualizing your data, you will go with a specific type of chart to portray your data.</p>

### Customization <br>
<p>There's many customizable options when using Matplotlib, such as; line types, colors, labels, figure size, etc. We'll go over a few of them here:</p>

##### Axis Labels & Line Marks

```python
# fmt = ['marker''line''color'] or ['color''marker''line']
plt.plot([1, 2, 3, 4, 5], [23, 45, 67, 78, 89], '*:m')
plt.plot([3, 4, 5, 6, 7], [22, 33, 44, 55, 66], 'c+--')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```

##### Legend & Line Labels

```python
plt.plot([1600, 1700, 1800, 1900, 2000], [1, 1.4, 1.6, 1.8, 2], 'o-r', label='Red Demographic')
plt.plot([1600, 1700, 1800, 1900, 2000], [2, 1.9, 2.8, 3, 2.4], 'x:g', label='Green Demographic')
plt.xlabel('Year')
plt.ylabel('Population (Millions)')
plt.legend()
plt.show()
```

### Subplots <br>
<p>Using the subplots method, you're able to create multiple plots within a single cell.</p>

```python
# first plot
plot_1 = plt.subplot(2, 1, 1) # Grab the Figure, Create a 2 rows by 1 column Axes, position plot_1 on the 1st (1st row, 1st column)
plt.plot([1600, 1700, 1800, 1900, 2000], [1, 1.4, 1.6, 1.8, 2], 'o-r', label='Red Demographic')
plt.xlabel('Year')
plt.ylabel('Population (Millions)')
plt.legend()

# second plot
plot_2 = plt.subplot(2, 1, 2)
plt.plot([1600, 1700, 1800, 1900, 2000], [2, 1.9, 2.8, 3, 2.4], 'x:g', label='Green Demographic')
plt.xlabel('Year')
plt.ylabel('Population (Millions)')
plt.legend()

plt.show()
```

### Presentation Worthiness
<p>Before we go into how to use this as a tool for data exploration, let's talk quickly about presentation-worthiness. In the next cell we're going to introduce customization of lines, colors, ticks, titles, fonts, and grids.</p>

```python
x = np.linspace(1,20, num=20)
y = x * 2
```

```python
fig, ax = plt.subplots(figsize = (20, 10))
x = np.linspace(1,20, num=20)
y = x * 2
plt.plot(x, y,
         lw=5, # line width
         ls='--', # line style
         color='r', # line color
         marker='*', # pointer marker
         ms=15, # marker size
         mfc='w', # marker face color
         mec='b', # marker edge color
         label='Data'
        )

plt.title('Visual of y = f(x)', fontsize=32, fontname='Times New Roman')

plt.xlabel('X Value', fontsize=24, fontname='Times New Roman')
plt.ylabel('Y Value', fontsize=24, fontname='Times New Roman')

plt.xticks(x, fontsize=24)
plt.yticks([y for y in range(0, 42, 4)], fontsize=24)

plt.grid(True)

plt.legend(loc='upper left', fontsize=24)
plt.show()
```

<p>It's sometimes nice to obsess a little bit over how your plots look. It's good practice paying attention to detail and it has huge payoff when you end up presenting your findings to other analysts and business partners. Here's a more complex example of using the matplotlib api:</p>

```python
markers = ['o', 'v', '^', '*', 's']
styles = ['-', '--', '-.', ':', '-']
colors = ['#588C7E', '#F2E394', '#F2AE72', '#D96459', '#8C4646']
```

```python
for i in range(5):
    print(i, markers[i], styles[i], colors[i])
```

```python
x = np.linspace(0, 10, 50)

fig, ax = plt.subplots(figsize=(20,10))

# loop 5 times to create 5 different plots on this figure
for i in range(5):
    plt.plot(x, x** i, 
             lw=5,
             ls=styles[i],
             c=colors[i],
             marker=markers[i],
             mfc=colors[-1-i],
             ms=15,
             label=f"x^{i}"
            )

plt.ylim(0,100)

plt.title('y = f(x^i) ', fontsize=32, fontfamily='serif', y=1.05)

plt.legend(fontsize=16)
plt.grid(True)
plt.show()
```

```python
import csv
from  itertools import groupby




colors = {'Iris-setosa': '#2B5B84', # Dark Blue/Purplish
          'Iris-versicolor': 'g',
          'Iris-virginica': 'purple'
         }
```

```python
colors = {
    'Iris-setosa': '#2B5B84', # Dark Blue/Purplish
    'Iris-versicolor': 'g',
    'Iris-virginica': 'purple'
}
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
flowers = pd.read_csv('iris.csv', sep=',', header=None, names=column_names)
flowers
```

```python
fig, ax = plt.subplots(figsize=(10,8))

for species, color in colors.items():
    flower_species = flowers[flowers['species'] == species]
    sepal_length = flower_species['sepal_length']
    sepal_width = flower_species['sepal_width']
    plt.scatter(sepal_length, sepal_width, c=color, label=species)
    

plt.title('Iris Dataset Length x Width', fontsize=20)
plt.xlabel('Sepal Length (cm)', fontsize=16)
plt.ylabel('Sepal Width (cm)', fontsize=16)
plt.grid(True)
plt.legend()
plt.show()
```

### Box Plot

```python
figure = plt.figure(figsize=(15, 10))

petal_lengths = []
for species in colors:
    species_petal_length = flowers[flowers['species'] == species]['petal_length']
    petal_lengths.append(species_petal_length)
    
plt.boxplot(petal_lengths)

plt.title('Box Plot with Iris Dataset', fontsize=16)
plt.xlabel('Flower Variety')
plt.ylabel('Petal Length (cm)')

plt.xticks([1, 2, 3], ['Setosa', 'Versicolor', 'Virginica'])

plt.yticks([num * .5 for num in range(15)])
plt.grid(True)

figure.savefig('iris_boxplot.pdf')
plt.show()
```

### Histograms <br>
<p>While line plots are great for visualizing trends in time series data, histograms are the king of visualizing distributions. Often, the distribution of a variable is what you're interested in, and a visualization provides a lot more information than a group of summary statistics.</p>

```python
fig, ax = plt.subplots(figsize=(16, 9)) #Subplots makes a call in the background to create a plot area

# generate random data
data = np.random.uniform(0, 10, 100)

plt.hist(data, bins=20 ,range=(0,10), label='Data', color='r', alpha=.75)

plt.xticks(list(range(0,11)))
plt.legend()
plt.show()
```

<p>Let's take a look at a mystery/randomly generated DataFrame, and plot the results</p>

```python
import pandas as pd
import numpy as np


def make_mystery_dataframe():
    customer_id = np.random.randint(1000,1200,10000)
    states = {1: 'California', 0: 'Colorado'}
    statelist = []
    agelist = []
    for i in range(10000):
        ind = np.random.randint(0,2)
        if ind == 0:
            statelist.append(states[ind])
            agelist.append(round(np.random.normal(loc = 30, scale = 3)))
        elif ind == 1:
            statelist.append(states[ind])
            agelist.append(round(np.random.normal(loc = 40, scale = 5)))
    mystery_df = pd.DataFrame.from_dict({
        'customer_id': customer_id,
        'age':agelist,
        'state': statelist
    })
    
    return mystery_df


df = make_mystery_dataframe()

df
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

# first subplot - histogram for Colorado
ax1.hist(df[df['state'] == 'Colorado']['age'], bins=20, label='Ages in Colorado')
ax1.set_xlabel('Age', fontsize=24)
ax1.set_ylabel('Frequency', fontsize=24)
ax1.legend()
ax1.set_title('Distribution of Ages in Colorado', fontsize=24)
ax1.set_xlim(0,60)

# Second subplot - histogram for California
ax2.hist(df[df['state'] == 'California']['age'], bins=20, color='r', label='Ages in California')
ax2.set_xlabel('Age', fontsize=24)
ax2.set_ylabel('Frequency', fontsize=24)
ax2.legend()
ax2.set_title('Distribution of Ages in California', fontsize=24)
ax2.set_xlim(0,60)

plt.show()
```
